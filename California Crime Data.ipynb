```python
print("\n8. What is the percentage of different crimes in the California?\n")  

crime_list=['Mur_mans','For_rape','Robbery','Agg_assault','Burglary','Lar_theft', 'Moto_theft', 'Arson']
dict_total = df[crime_list].sum().to_dict()
dt=pd.DataFrame.from_dict(dict_total, orient='index',columns=['Total'])
dt['Percentage %'] = dt.Total * 100 / dt.Total.sum()
dt
```

```python
print("\n\n9. Which is the most frequently occurring crime in California?")

dt[dt.Total == dt.Total.max()].index[0]
```

```python
print("\n\n10. Which crime is frequently occurring when the population is less than 5, 00,000? Create a frequency distribution table where classes (in thousands) are 0-500, 500-1000, etc.")
df2 = df[df.Popul<500000]
total_crime = df2[crime_list].sum().to_dict()
dt2=pd.DataFrame.from_dict(total_crime, orient='index',columns=['Total'])
dt2['Percentage %'] = dt.Total * 100 / dt.Total.sum()
print(f"\n\n Frequently occuring crime with popul < 5, 00,000 is {dt2[dt2.Total == dt2.Total.max()].index[0]}")
```

```python
print("\n\nFrequency Distribution Table")
dict1 = pd.cut(df.Mur_mans, bins=range(0, df.Mur_mans.max()+500, 500)).value_counts().to_dict()
dict2 = pd.cut(df.For_rape, bins=range(0, df.For_rape.max()+500, 500)).value_counts().to_dict()
dict3 = pd.cut(df.Robbery, bins=range(0, df.Robbery.max()+500, 500)).value_counts().to_dict()
dict4 = pd.cut(df.Agg_assault, bins=range(0, df.Agg_assault.max()+500, 500)).value_counts().to_dict()
dict5 = pd.cut(df.Burglary, bins=range(0, df.Burglary.max()+500, 500)).value_counts().to_dict()
dict6 = pd.cut(df.Lar_theft, bins=range(0, df.Lar_theft.max()+500, 500)).value_counts().to_dict()
dict7 = pd.cut(df.Moto_theft, bins=range(0, df.Moto_theft.max()+500, 500)).value_counts().to_dict()
dict8 = pd.cut(df.Arson, bins=range(0, df.Arson.max()+500, 500)).value_counts().to_dict()

dict_use={'Mur_mans':dict1,'For_rape':dict2,'Robbery':dict3,'Agg_assault':dict4,'Burglary':dict5,'Lar_theft':dict6,
          'Moto_theft':dict7, 'Arson':dict8}

dtf=pd.DataFrame.from_dict(dict_use, orient='columns').fillna(0)
dtf=dtf.transpose()
dtf
dtf.loc[:, (dtf != 0).any(axis=0)]
```
