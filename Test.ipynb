### Create a heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
questions = ["How are you?", "What can you do?", "Where?", "Why is it blue?"]
df = pd.DataFrame({'bot1': [0, 1, 0, 0],
                    'bot2': [0, 1, 1, 0],
                   'bot3': [1, 1, -1, 1]}, index=questions)
```

```python
sns.set_style("dark")
sns.heatmap(df)
plt.show()
```
